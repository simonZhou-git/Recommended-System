```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
columns = ['user_id', 'item_id', 'rating', 'timestamp']
frame = pd.read_csv('u.data', sep = '\t', names = columns)
frame.head()
```

```python
columns = ['item_id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
          'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
          'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

movies = pd.read_csv('u.item', sep = '|', names = columns, encoding = 'latin-1')
movies.head()
```

```python
combined = pd.merge(frame, movies, on = 'item_id')
combined.head()
```

```python
combined.groupby('item_id')['rating'].count().sort_values(ascending = False).head()
```

```python
combined[combined['item_id'] == 50]['movie title'].unique()
```

```python
rating_crosstab = combined.pivot_table(values = 'rating', index = 'user_id', columns = 'movie title', fill_value = 0)
rating_crosstab.head()
```

```python
rating_crosstab.shape
```

```python
transpose = rating_crosstab.values.T
transpose.shape
```

```python
SVD = TruncatedSVD(n_components = 12, random_state = 17)

matrix = SVD.fit_transform(transpose)
matrix.shape
```

```python
corr_matrix = np.corrcoef(matrix)
corr_matrix.shape
```

```python
movie_names = rating_crosstab.columns
movie_names
```

```python
list_movie = list(movie_names)
starwar = list_movie.index('Star Wars (1977)')
starwar
```

```python
corr_wars = corr_matrix[starwar]
corr_wars.shape
```

```python
list(movie_names[(corr_wars < 1.0) & (corr_wars > 0.9)])
```

```python
list(movie_names[(corr_wars < 1.0) & (corr_wars > 0.95)])
```
