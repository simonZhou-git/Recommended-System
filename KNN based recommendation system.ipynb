```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.neighbors import NearestNeighbors
```

```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars.head()
```

```python
cars.rename(columns = {'Unnamed: 0': 'cars_name'}, inplace = True)
cars.head()
```

```python
target = [15,300,160,3.2]

x = cars.iloc[:, [1,3,4,6]].values
x
```

```python
nbrs = NearestNeighbors(n_neighbors = 1).fit(x)
print(nbrs.kneighbors([target]))
```
